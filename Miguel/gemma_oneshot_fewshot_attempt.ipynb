# Generic functions to attain the data

```python
def save_data():
    url = "https://github.com/rasbt/python-machine-learning-book-3rd-edition/raw/master/ch08/movie_data.csv.gz"
    filename = url.split("/")[-1]

    with open(filename, "wb") as f:
        r = requests.get(url)
        f.write(r.content)

    with gzip.open(filename, 'rb') as f_in:
        with open('movie_data.csv', 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
```

```python
def prepare_data_test(df):
    X_test = df.iloc[40000:42500]
    X_test['text'] = X_test['review']
    X_test['text'] = X_test['text']
    y_true = list(X_test['sentiment'])
    return X_test.drop(['sentiment','review'],axis=1), y_true
```

```python
extras = None
def prepare_data():
    global extras
    df = pd.read_csv('movie_data.csv')
    extras = df.iloc[0:1000]
    return prepare_data_test(df)
save_data()
X_test, y_true = prepare_data()
```

```python
print(extras)
```

```python
entries_with_value_1 = extras[extras.iloc[:, -1] == 1].sample(n=20, random_state=42)
entries_with_value_0 = extras[extras.iloc[:, -1] == 0].sample(n=20, random_state=42)
print(entries_with_value_1)
print(entries_with_value_0)
```

```python
reviews_with_value_1 = entries_with_value_1['review'].tolist()
reviews_with_value_0 = entries_with_value_0['review'].tolist()
```

# Zero Shot Text Classification

Pipeline preparation

```python
from transformers import pipeline
from tqdm import tqdm

classifier = pipeline("zero-shot-classification", model="google/gemma-2b", device=0)

labels = ["positive", "negative"]

classifications = []
```

# Zero Shot Learning

```python
for review in tqdm(X_test['text']):
    #print(review)
    
    classification = classifier(review, labels)
    classifications.append(classification)
    
    #print(classification)

# Now 'classifications' contains the predicted sentiment labels (0 or 1) for each element in x_test
# print(classifications)
```

# One Shot Learning

Here we utilize the same settings for the pipeline, but to do one shot learning we must give out one example of a positive and one example of a negative classification.

As noticeable bellow we could not run this code as it exausted the available resources.

```python
from transformers import pipeline
from tqdm import tqdm

classifier = pipeline("zero-shot-classification", model="google/gemma-2b", device=0)

labels = ["positive", "negative"]

classifications = []

one_shot_positive = reviews_with_value_1[0]
one_shot_negative = reviews_with_value_0[0]

for review in tqdm(X_test['text']):
    classification = classifier(review, labels, positive_examples=[one_shot_positive], negative_examples=[one_shot_negative])
    classifications.append(classification)

# Now 'classifications' contains the predicted sentiment labels for each element in X_test
#print(classifications)
```

# Few Shot Learning

Here we utilize the same settings for the pipeline, but to perform few shot learning we must give out multiple examples of positive and negative reviews.

As noticeable bellow we could not run this code as it exausted the available resources.

```python
from transformers import pipeline
from tqdm import tqdm

classifier = pipeline("zero-shot-classification", model="google/gemma-2b", device=0)

labels = ["positive", "negative"]

classifications = []

for review in tqdm(X_test['text']):
    classification = classifier(review, labels, positive_examples=reviews_with_value_1, negative_examples=reviews_with_value_0)
    classifications.append(classification)

# Now 'classifications' contains the predicted sentiment labels for each element in X_test
#print(classifications)
```

# Result analysis

```python
results = []

for classification in classifications:
    if classification['labels'][0] == "positive":
        results.append(1)
    else:
        results.append(0)

print(results)
```

```python
# Calculate accuracy
correct_predictions = sum(1 for pred, true in zip(results, y_true) if pred == true)
total_predictions = len(y_true)
accuracy = correct_predictions / total_predictions

print("Accuracy:", accuracy*100,"%")
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_true, results)

print("Classification Report:")
print(report)
```
